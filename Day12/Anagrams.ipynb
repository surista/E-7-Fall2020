```python
def are_anagrams(word1, word2):
    '''Judge whether two words are anagrams'''
    
    # Words should be lower cases
    word1 = word1.lower()
    word2 = word2.lower()
    
    # The same words are not anagrams
    if word1 == word2:
        return False

    # All the characters in one word should be in the other
    for ch in word1:
        if not ch in word2:
            return False
        
    # The two words should have the same number of each characters
    for ch in word1:
        if word1.count(ch) != word2.count(ch):
            return False
        
    # Two words with the same length are anagrams if the requirements above are met
    return True
```

```python
# anagrams.py
#
# This program takes a text file and finds all of the sets of anagrams contained within the text.
#
# Usage:
#   % python anagrams.py yourfile.txt
 

# Define a function that finds anagram sets in a text file
def find_anasets(text):
    text_file = open(text)
    # Open a dictionary that will eventually contain sets of anagrams
    words = {}
    # Loop through each word in the provided text file, separating each word on white space
    for word in text_file.read().split():
        # Convert each word to lower case, for easier comparison
        word2 = word.lower()
        # Create a variable containing the "original" word we are analyzing
        wordsave = word2
        # Create a list that will contain the characters of each word
        l1 = []
        # Loop through each character in the word, creating a list of the characters
        for ch in word2:
            l1.append(ch)
        # Sort this list in alphabetical order and join together to form one string
        word2 = ''.join(sorted(l1))
        # The main idea is that words that are anagrams will have this same sorted "word2" string from above. Thus, if
        # word2 is not in the keys in our dictionary, we create a new dictionary entry with words2 as the key and
        # wordsave as the entry. In the future, we may find other words with the same word2, in which case this would be
        # an anagram
        if word2 not in words.keys():
            words[word2] = [wordsave]
            continue
        # If word2, the alphabetically ordered characters of our word in question, is already in the dictionary, this
        # means we have found an anagram. Thus, we append wordsave (the actual word) to the list associated with
        # the word2 key, builing out a list of all the words that have the same word2 and thus are anagrams of one
        # another
        else:
            words[word2].append(wordsave)
        # looping through this process for each word in the list will get us a dictionary where all anagrams are grouped
        # together under one key in the dictionary
    # Since we want to rank our anagram sets, we will need to get them into a list
    out = []
    # Loop through all the keys and values in the dictionary, appending each dictionary value as a new sublist in our
    # out list. However, I first covert each of the dictionary values into a set in order to eliminate any duplicate
    # entries (lists only contain distinct values), since an anagram should not be counted as an anagram of itself.
    for keys, values in words.items():
        nodoubles = set(values)
        if len(nodoubles) > 1:
            out.append(list(nodoubles))
    # Sort our list of lists by reverse length, such that the longest lists are those that appear first
    final = sorted(out, key=len, reverse=True)
    # Print all of the entries in list "out", including the length of the sublist and the sublist itself.
    for group in final:
        print((len(group), group))
    # If we wanted to just print the 10 longest items from the list (as Q1 requires), we can just add [:10] after 'in
    # final:' above
    for group in final[:10]:
        print((len(group), group))

import sys
import time

# capture the time at the start of the program to track how long executing the function takes
start = time.time()
# error response if the user forgets to include the command line argument speficying what file to use. If sys.argv
# does not have an argument in the first position (len <2), then the user has forgotten to specify a file to use and
# we return the appropriate error.
if len(sys.argv) < 2:
    print("No file specified. Please specify a file to search for interlacing words.")
    sys.exit()
else:
    # assuming that the file for use is specified, bring in the file name from the comand prompt
    filename = sys.argv[1]
    # we try inputting the filename provided on the command line into the find_anasets function.
    try:
        find_anasets(filename)
    # we use an exception to deal with the circumstance where python returns a FileNotFoundError, meaning the file was
    # specified but could not be located in the directory.
    except FileNotFoundError:
        print("File " + filename + " not found.")
    # for any other errors returned from running find_anasets with our filename, we use another exception that returns
    # a "could not read file" error message"
    except:
        print("Could not read the file")

# to find out how long the program took to execute, we mark the end time of the program and subtract the start time
end = time.time()
print(end-start)
```

```python
# anagrams.py
#
# This program takes a text file and finds all of the sets of anagrams contained within the text.
#
# Usage:
#   % python anagrams.py yourfile.txt


# finds sets of anagram of words in a text file
def find_anasets(text):

    text_file = open(text)
    # Dictionary to contain sets of anagrams
    words = {}

    # For each word in the text file
    for word in text_file.read().split():
        # Convert to lower case for comparison
        word = word.lower()

        # Create a list that will contain the characters of each word
        letters = list(word)

        # Sort the letters 
        representative = ''.join(sorted(letters))

        # All anagrams share the same sorted string 
        if representative not in words.keys():
            words[representative] = [word]
        else:
            words[representative].append(word)

    # Rank our sets of anagram 
    out = []

    # Loop through all the keys and values in the dictionary
    for keys, values in words.items():
        if len(values) > 1:
            out.append(values)

    # Sort our list so of longest lists appear first
    final = sorted(out, key=len, reverse=True)

    # Print the 10 longest items from the list 
    for group in final[:10]:
        print((len(group), group))

import sys
import time

# capture start time 
start = time.time()

if len(sys.argv) < 2:
    print("No file specified. Please specify a file to search for interlacing words.")
    sys.exit()
else:    
    try:
        find_anasets(sys.argv[1])
    except FileNotFoundError:
        print("File " + sys.argv[1] + " not found.")
    except:
        print("Could not read the file")

# mark the end time of the program 
end = time.time()

# Find difference
print(end - start)
```
